```python
import requests
import re
import os
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlsplit

nd_list = []
for url in list:
    nd_dict = {}
    driver.get(url)
    sleep(1)
    element1= WebDriverWait(driver, 90).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div[class="styles__StyledProductContent-sc-1f8f774-0 ewqXRk"]')))
    html_of_interest1=driver.execute_script('return arguments[0].innerHTML',element1)
    soup1=BeautifulSoup(html_of_interest1, 'html.parser')
    Ten_sach = soup1.select('h1[class="title"]')[0].text.strip()
    
    if soup1.select('h6') != []:
        Tac_gia = soup1.select('h6')[0].select('a')[0].text.strip()
    else:
        Tac_gia = ''
    
    if soup1.select('div[class="product-price has-discount"]') != []:
        Gia = soup1.select('div[class="product-price has-discount"]')[0].select('div')
        Gia_ban = Gia[0].text.strip()
        Gia_goc = Gia[1].text.strip()
        Giam_gia = Gia[2].text.strip()
    elif soup1.select('div[class="product-price"]') != []:
        Gia = soup1.select('div[class="product-price"]')[0].select('div')
        Gia_ban = Gia[0].text.strip()
        Gia_goc = ''
        Giam_gia = ''
    elif soup1.select('div[class="flash-sale-price"]') != []:
        Gia = soup1.select('div[class="flash-sale-price"]')[0].select('span')
        Gia_ban = Gia[0].text.strip()
        Gia_goc = Gia[1].text.strip()
        Giam_gia = Gia[2].text.strip()
    else:
        Gia = soup1.select('div[class="styles__Container-sc-6hj7z9-0 dFqfkL"]')[0].select('div')
        Gia_ban = Gia[0].text.strip()
        Gia_goc = Gia[2].text.strip()
        Giam_gia = Gia[3].text.strip() 
    
    if soup1.select('div[class="styles__StyledQuantitySold-sc-1u1gph7-2 exWbxD"]') != []:
        So_luong_ban = soup1.select('div[class="styles__StyledQuantitySold-sc-1u1gph7-2 exWbxD"]')[0].text.strip()
    else:
        So_luong_ban = ''

    if soup1.select('a[class="number"]') != []:
        click = soup1.select('a[class="number"]')
    else:
        click = []

    element2= WebDriverWait(driver, 90).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div[class="style__Wrapper-sc-12gwspu-0 cIWQHl"]')))
    html_of_interest2=driver.execute_script('return arguments[0].innerHTML',element2)
    soup2=BeautifulSoup(html_of_interest2, 'html.parser')
    # if soup2.select('div[class="content has-table"]') != []:
    #     ttct = soup2.select('div[class="content has-table"]')[0].select('td')
    #     d = 0
    #     Thong_tin_chi_tiet = ''
    #     for i in range(len(ttct)):
    #         d += 1
    #         if d % 2 != 0:
    #             Thong_tin_chi_tiet += ttct[i].text + ': '
    #         else:
    #             Thong_tin_chi_tiet += ttct[i].text + '\n'
    #     Thong_tin_chi_tiet = Thong_tin_chi_tiet.strip()
    # else:
    #     Thong_tin_chi_tiet = ''

    Mo_ta_san_pham = soup2.select('div[class="ToggleContent__View-sc-1dbmfaw-0 wyACs"]')[0].text.strip()
    
    element3= WebDriverWait(driver, 90).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div[data-view-id="breadcrumb_container"]')))
    html_of_interest3=driver.execute_script('return arguments[0].innerHTML',element3)
    soup3=BeautifulSoup(html_of_interest3, 'html.parser')
    The_loai = ''
    for i in range(3,len(soup3.select('a[class="breadcrumb-item"]'))-1):
        The_loai += soup3.select('a[class="breadcrumb-item"]')[i].text.strip() + '\n'
    The_loai = The_loai.strip()
        
    if click != []:
        driver.execute_script("arguments[0].click();", WebDriverWait(driver, 90).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'a[class="number"]'))))
        element4= WebDriverWait(driver, 90).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div[class="review-rating__inner"]')))
        html_of_interest4=driver.execute_script('return arguments[0].innerHTML',element4)
        soup4=BeautifulSoup(html_of_interest4, 'html.parser')
        Diem_danh_gia = soup4.select('div[class="review-rating__point"]')[0].text.strip()
        So_luong_nhan_xet = soup4.select('div[class="review-rating__total"]')[0].text.strip()
    else:
        Diem_danh_gia = ''
        So_luong_nhan_xet = ''
    
    nd_dict['link'] = url
    nd_dict['Tên sách'] = Ten_sach
    nd_dict['Tác giả'] = Tac_gia
    nd_dict['Thể loại'] = The_loai
    nd_dict['Giá bán'] = Gia_ban
    nd_dict['Giá gốc'] = Gia_goc
    nd_dict['Giảm giá'] = Giam_gia
    nd_dict['Số lượng bán'] = So_luong_ban
    # nd_dict['Thông tin chi tiết'] = Thong_tin_chi_tiet
    nd_dict['Mô tả sản phẩm'] = Mo_ta_san_pham
    nd_dict['Điểm đánh giá'] = Diem_danh_gia
    nd_dict['Số nhận xét'] = So_luong_nhan_xet
    
    nd_list.append(nd_dict)

df = pd.DataFrame(nd_list)
df
```
